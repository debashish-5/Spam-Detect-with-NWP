```python
df['spam'] = df['category'].apply(lambda x: 1 if x == 'spam' else 0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_set,test_set = train_test_split(df,test_size=0.2,random_state=42)
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.message,df.spam,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
x_train_num = cv.fit_transform(x_train)
```

```python
x_train_num
```

```python
x_train_num.toarray()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dec = DecisionTreeClassifier()
```

```python
dec.fit(x_train_num,y_train)
```

```python
input_user = input('Enter the Message:')
input_user_num = cv.transform([input_user])
predition = dec.predict(input_user_num)
if predition == 1:
    print('spam message')
else:
    print('not spam message')
```
